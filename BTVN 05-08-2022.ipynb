```python
# Bài 1: (Dễ): Nhập vào một chuỗi gồm các số tự nhiên cách nhau một khoảng trắng. 
#         Tính tổng, tính trung bình cộng các số. 
#         Tìm số âm lớn nhất và số dương bé nhất trong chuỗi vừa nhập.

#1 2 3 4 8 5 6
#               -5 -2 -1 -3 -4 8 9 7 5 7 8 4 3 -6

string = input('nhập chuỗi: ')
string = string.strip()
string1 = string.split()
string2 = string.split()
string3 = string.split()
x = len(string1)
tong = 0
n = 0
for i in range(x):
    b = int(string1[i])
    tong += b
    n+=1

TB = tong/n

for i in range(x):
    a1 = int(string1[i])
    if a1 < 0: 
        string2.append(a1)
        x2 = len(string2[x:])
        str2 = (string2[x:])

    if a1 > 0: 
        string3.append(a1)
        x3 = len(string3[x:])
        str3 = (string3[x:])

        
max_ = str2[1]
for i in range(x2):        
    a = str2[i]
    b = str2[1]
    if a > max_: 
        max_ = a
    if max_ == b:
        max_ = b
    else:
        max_ = max_   
        
min_ = str3[1]
for j in range(x3):        
    a3 = str3[j]
    b3 = str3[1]
    if a3 < min_: 
        min_ = a3
    if min_ == b3:
        min_ = b3
    else:
        min_ = min_
         
print ('\ntổng là: ',tong)
print ('trung bình là: ', TB)
print ('số nguyên âm lớn nhất là: ',max_)
print ('số nguyên dương nhỏ nhất là: ', min_)
```

```python
import turtle
import math

t = turtle.Turtle()
t.penup()
t.goto(-250, -300)
t.pendown()

def goto(x,y):
    t.penup()
    t.goto(x, y)
    t.pendown()
#vẽ đường ngang
for i in range(2):
    t.forward(560)
    t.left(90)
    t.forward(630)
    t.left(90)
goto((-250-10),(-300-10))
t.pensize(3.5)
for i in range(2):
    t.forward(580)
    t.left(90)
    t.forward(650)
    t.left(90)
t.pensize(1)

#viết số
# t.right(90)
# for i in range(9,0,1):
#     goto((-250+(70*(9-i)),(-300-15))
#     t.write(i, True, align = "center")
    
for i in range(9):
    t.penup()
    t.goto(-250, (-300+(70*i)))
    t.pendown()
    t.forward(560)

goto(-250, -300)

#vẽ đường dọc
t.left(90)
for i in range(8):
    t.penup()
    t.goto((-250+(70*i)), (-300))
    t.pendown()
    t.forward(280)

t.left(180)
goto(-250, (-300+630))
for i in range(8):
    t.penup()
    t.goto((-250+(70*i)), (-300+630))
    t.pendown()
    t.forward(280)

t.left(90)
goto((-250+210),-300)
t.goto((-250+350),(-300+140))
goto((-250+350),-300)
t.goto((-250+210),(-300+140))

goto((-250+210), (-300+630))
t.goto((-250+350), (-300+490))
goto((-250+350), (-300+630))
t.goto((-250+210), (-300+490))

def saophaidown(x, y):
    goto(x+3,y-10)
    t.forward(7)
    t.right(90)
    t.forward(7)
    
def saophaiup(x, y):
    goto(x+3,y+10)
    t.forward(7)
    t.left(90)
    t.forward(7)

def saotraidown(x, y):
    goto(x-3,y-10)
    t.forward(7)
    t.left(90)
    t.forward(7)
    
def saotraiup(x, y):
    goto(x-3,y+10)
    t.forward(7)
    t.right(90)
    t.forward(7)
def phai(x,y):
    goto((x+(70*i)), y)
    saophaidown((x+(70*i)), y)
    t.right(90)
    saophaiup((x+(70*i)), y)
    t.left(90)
def trai(x,y):
    goto((x-(70*i)),y)
    saotraidown((x-(70*i)),y)
    t.left(90)
    saotraiup((x-(70*i)), y)
    t.right(90)   
def trai1(x,y):
    goto((x+(70*i)),y)
    saotraidown((x+(70*i)),y)
    t.left(90)
    saotraiup((x+(70*i)), y)
    t.right(90)    

t.left(90)
#t hướng lên
goto(-250, (-300+210))
for i in range(0,7,2):
    phai(-250, -90)
#t hướng lên
goto(-250, (-300+420))
for i in range(0,7,2):
    phai(-250,120)
#t hướng lên

goto((-250+560), (-300+210))
for i in range(0,7,2):
    trai(310, -90)
#t hướng lên
goto((-250+560), (-300+420))
for i in range(0,7,2):
    trai(310,120)
#t.hướng lên

goto((-250+70), (-300+140))
for i in range(0,7,6):
    phai(-180,-160)
    trai1(-180,-160)
goto((-250+70),(-300+490))
for i in range(0,7,6):
    phai(-180, 190)
    trai1(-180, 190)
#t hướng lên
  
t.hideTurtle()
turtle.done()
    
```
